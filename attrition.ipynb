## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = [
    'Age', 'DistanceFromHome', 'Education', 'JobLevel', 'NumCompaniesWorked',
    'PercentSalaryHike', 'TotalWorkingYears', 'TrainingTimesLastYear',
    'YearsAtCompany', 'YearsSinceLastPromotion']


# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
y_df_encoded = pd.get_dummies(y_df['Attrition'], drop_first=True)  # Yes -> 1, No -> 0
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df_encoded, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_df_numeric = X_df.apply(pd.to_numeric, errors='coerce')
nan_values = X_df_numeric.isna().sum()
nan_values[nan_values > 0]
```

```python
X_df_numeric
```

```python
X_df_numeric.info()
```

```python
X_df_numeric.value_counts()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled.shape, X_test_scaled.shape
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')

# Fit the encoder to the training data
encoder.fit(y_df[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
department_train_encoded = encoder.transform(y_df.loc[X_train.index, ['Department']])
department_test_encoded = encoder.transform(y_df.loc[X_test.index, ['Department']])
department_train_encoded.shape, department_test_encoded.shape

```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse=False, drop='first')

# Fit the encoder to the training data
attrition_encoder.fit(y_df[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
attrition_train_encoded = attrition_encoder.transform(y_df.loc[X_train.index, ['Attrition']])
attrition_test_encoded = attrition_encoder.transform(y_df.loc[X_test.index, ['Attrition']])
attrition_train_encoded.shape, attrition_test_encoded.shape
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_columns = X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(num_columns,))

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
shared_layer2
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_input = layers.Input(shape=(2,))
# Create the hidden layer
department_hidden = layers.Dense(16, activation='relu')(department_input)

# Create the output layer
department_output = layers.Dense(1, activation='sigmoid')(department_hidden)
department_output
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_input = layers.Input(shape=(2,))
# Create the hidden layer
attrition_hidden = layers.Dense(16, activation='relu')(attrition_input)

# Create the output layer
attrition_output = layers.Dense(1, activation='sigmoid')(attrition_hidden)
attrition_output
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
combined = layers.concatenate([shared_layer2, department_hidden])
```

```python
# Create the model
final_output = layers.Dense(1, activation='sigmoid')(combined)
model = Model(inputs=[input_layer, department_input], outputs=final_output)


# Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])


# Summarize the model
model.summary()
```

```python
print(X_train_scaled.shape)            # Check the shape of X_train_scaled
print(department_train_encoded.shape)  # Check the shape of department_train_encoded
print(y_train.shape)                   # Check the shape of y_train
```

```python
y_train = y_train.values if isinstance(y_train, pd.Series) else y_train
y_test = y_test.values if isinstance(y_test, pd.Series) else y_test
```

```python
# Train the model
trained_model = model.fit(
    [X_train_scaled, department_train_encoded],
    y_train,
    epochs=50,
    batch_size=34,
    validation_data=([X_test_scaled, department_test_encoded], y_test),
    verbose=2
)
```

```python
# Evaluate the model with the testing data
test_loss, test_accuracy = model.evaluate([X_test_scaled, department_test_encoded], y_test, verbose=2)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Print the accuracy for both department and attrition
evaluation = model.evaluate([X_test_scaled, department_test_encoded], y_test, verbose=2)
print(f"Test Loss: {evaluation[0]}")
print(f"Test Accuracy (Attrition): {evaluation[1]}")
```
